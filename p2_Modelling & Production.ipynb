```python
import collections
import pandas as pd
#imports
```

```python
with open('test.txt', 'r') as info:
  count = collections.Counter(info.read())
  print(count.items())
  
```

```python
#final_df = pd.DataFrame(count)
df = pd.DataFrame.from_dict(count, orient='index').reset_index()
df
```

```python
df = df.rename(columns={'index':'Characters', 0:'Occurance'})
#df['Column'] = df['Column'].astype(float)
df['Occurance %']=df['Occurance']
df['Occurance %']=df['Occurance %'].astype(float)
x=df['Occurance'].sum ()
print(df['Occurance'].sum ())
for i in df.axes[0]:
 df['Occurance %'][i]=df['Occurance'][i]/x
df
```

```python
df.to_csv('out.csv')
```
